```python
#plotting Weights
#create a figure

plt.figure(figsize=(20, 5))
x = range(da['BMXWT'].size)
y = da['BMXWT']
plt.scatter(x, y, marker='x', s= 0.1)

#very important from stats prespective to make figure readable
plt.title("Weights of Individuals")
plt.xlabel("ID of Individuals")
plt.ylabel("Weight (KGs)")


plt.show()
```
