```python
#Studying Central Tendency
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from scipy import stats
import seaborn as sns
```

```python
url = "nhanes_2015_2016.csv"
da = pd.read_csv(url)
```

```python
plt.figure(figsize = (20, 5))
x = range(da['BMXWT'].size)
y = da['BMXWT']
plt.scatter(x, y, marker='x', s= 0.1)

#very important from stats prespective to make figure readable
plt.title("Weights of Individuals")
plt.xlabel("ID of Individuals")
plt.ylabel("Weight (KGs)")
p1 = (0, 75)
p2 = (6000, 75)
x_c, y_c = zip(p1, p2)
plt.plot(x_c, y_c, color = 'red', linewidth = 0.75)
plt.show()
```

```python
wt =  da['BMXWT']
```

```python
wt.mean()
```

```python
plt.figure(figsize = (20, 5))
x = range(da['BMXWT'].size)
y = da['BMXWT']
plt.scatter(x, y, marker='x', s= 0.1)

#very important from stats prespective to make figure readable
plt.title("Weights of Individuals")
plt.xlabel("ID of Individuals")
plt.ylabel("Weight (KGs)")

x_c, y_c = ([0, 6000], [75, 75])
plt.plot(x_c, y_c, color = 'red', linewidth = 0.75)

x_c, y_c = ([0, 6000], [wt.mean(), wt.mean()])
plt.plot(x_c, y_c, color = 'green', linewidth = 2)

plt.show()
```

```python
plt.figure(figsize = (20, 5))
x = range(da['BMXWT'].size)
y = da['BMXWT']
plt.scatter(x, y, marker='x', s= 0.1)

#very important from stats prespective to make figure readable
plt.title("Weights of Individuals")
plt.xlabel("ID of Individuals")
plt.ylabel("Weight (KGs)")

x_c, y_c = ([0, 6000], [wt.median(), wt.median()])
plt.plot(x_c, y_c, color = 'red', linewidth = 2)

x_c, y_c = ([0, 6000], [wt.mean(), wt.mean()])
plt.plot(x_c, y_c, color = 'green', linewidth = 2)

plt.show()
```
